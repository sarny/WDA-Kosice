### Úloha 3.5

Vytvorte zoznam `indexy` do ktorého v cykle priradíte indexy prvkov zoznamu `cisla`, ktoré majú hodnoty menšie než 45. Skúste úlohu riešiť najrpv pomocou cyklu `while` a potom pomocou `for`.

```python
cisla = [25, 24, 26, 45, 25, 23, 50, 51]

# zoznam 'indexy' by mal byť rovný [0, 1, 2, 4, 5]
indexy = []

# riešenie pomocou while
i = 0 # inicializujeme si pomocnú premennu pre indexovanie
while i < len(cisla):
    cislo = cisla[i] # získame hodnotu prvku na i-tej pozícii
    if cislo < 45:
        indexy.append(i) # ak hodnota spĺňa podmienku, pridáme jej index do výsledného zoznamu indexes
    i += 1 # nesmieme na konci zabudnúť inkrementovať indexovú premennu
    
print(indexy)

# riešenie pomocou for a enumerate
indexy = []
for i, cislo in enumerate(cisla):
    if cislo < 45:
        indexy.append(i)

print(indexy)

# riešenie pomocou for a indexovania vygenerovaného pomocou range
indexy = []
for i in range(len(cisla)): # vygenerujeme si postupnosť indexov od 0 do (počet prvkov zoznamu cisla - 1)
    cislo = cisla[i] # získame hodnotu prvku na i-tej pozícii
    if cislo < 45:
        indexy.append(i)

print(indexy)
```

### Úloha 3.9
Naprogramujte skript jednoduchej kalkulačky, ktorá v cykle načíta na vstupe reťazec reprezentujúci číselný výraz v tvare:
`[číslo1] [operácia] [číslo2]` a vypočíta jeho číselnú hodnotu. Čísla môžu byť celé, alebo desatinné, a operácia môže byť +-*/.
Výsledok vypíšte na 4 desatinné miesta.

```python
# ako vstup pre testovanie použite nasledujúci zoznam výrazov
vyrazy = ["1 + 2", "2 * 3.14", "2 - 3.0", "10 / 5"]

# v cykle pre každý výraz
for vyraz in vyrazy:
    operandy = vyraz.split(" ") # rozdelíme si výraz na jednotlivé časti (operandy)
    
    x = operandy[0]   # prvé číslo
    opr = operandy[1] # operácia
    y = operandy[2]   # druhé číslo
    
    # prevedieme reťazce na čísla
    x = float(x) if x.find(".") >= 0 else int(x)  # ak reťazec obsahuje desatinnú bodku . skonvertujeme ho na float (desatinné číslo), inak na int (celé číslo)
    y = float(y) if y.find(".") >= 0 else int(y)
    
    if opr == "+":
        vysledok = x + y
    elif opr == "-":
        vysledok = x - y
    elif opr == "*":
        vysledok = x * y 
    elif opr == "/":
        vysledok = x / y
        
    print("{0:.4f}".format(vysledok))
```

### Úloha 3.10
Na vstupe máte nasledujúce textové dáta:

```python
data = """'power'; 'weight'; 'hybrid'; 'producer'
55.3; 1450; false; 'ford'
100; 2500; false; 'ford'
60.5; 1150; true; 'toyota'
56; ?; false; 'volkswagen'
75.2; 1480; false; 'volkswagen'
80.1; 1630; false; 'peugeot'"""
```

Napíšte skript, ktorý:
* Rozdelí text na riadky a oddelí dátové polia. Hodnoty polí prevedie z reťazca na dátový typ Pythonu, tzn. čísla, reťazce bez oddeľovačov a Boolovské hodnoty.
* Chýbajúce hodnoty označené `?` reprezentujte konštantou `None`.
* Pre číselné polia vypíše základné štatistiky: min./max./priemernú hodnotu zaokrúhlené na 4 desatinné miesta.
* Pre textové polia vypíše zoznam hodnôt usporiadaný podľa abecedy.

```python
# najprv si text rozdelíme na riadky a každý riadok rozdelíme na hodnoty
retazce = data.split("\n")
riadky = []
for riadok in retazce:
    riadky.append(riadok.split(';'))

# skonvertujeme hodnoty z reťazcov na typy Pythonu
for riadok in riadky:
    for i, hodnota in enumerate(riadok): # budeme chcieť zmeniť hodnotu, takže budeme potrebovať jej index
        # najprv odstránime z hodnoty prázdne znaky zo začiatku a konca
        hodnota = hodnota.strip()
        if hodnota == "?":  # najprv skonvertujeme jednoduché hodnoty None, True a False
            hodnota = None
        elif hodnota == "true":
            hodnota = True
        elif hodnota == "false":
            hodnota = False
        elif hodnota.startswith("'"): # ak je hodnota ohraničená '', tak je to reťazec
            hodnota = hodnota[1:-1]   # odstránime z neho ohraničujúce znaky (prvý a posledný)
        elif hodnota.find(".") >= 0:  # inak je to číslo, ak obsahuje desatinnú bodku, prevedieme ho na float, inak na int
            hodnota = float(hodnota)
        else:
            hodnota = int(hodnota)
        riadok[i] = hodnota # zmenenú hodnotu prepíšeme v riadku

# pre kontrolu si zobrazíme skonvertované dáta (print zobrazí aj ohraničenie Python reťazcov '', ale hodnoty už neobsahujú ')
print(riadky)

# z rows odstránime prvý riadok (hlavičku tabuľky s názvami stĺpcov)
del(riadky[0])

# v jednom cykle vypočítame min, max a priemernú hodnotu pre prvý stĺpec (power)
suma = 0
min = None
max = None
for riadok in riadky:
    hodnota = riadok[0]
    if hodnota is None: # preskočíme chýbajúce hodnoty  
        continue
    if min is None or hodnota < min:  # ak sme ešte nepriradili do min žiadnu hodnotu, alebo ak je hodnota v riadku menšia než minimum, priradíme ju do min
        min = hodnota
    if max is None or hodnota > max:  # podobne ak sme ešte nepriradili max alebo je hodnota väčšia než max, priradíme ju do max 
        max = hodnota
    suma += hodnota # pripočítame hodnotu do sumy pre priemer
priemer = suma / len(riadky)
    
print("power min: {0:.4f}, max: {1:.4f}, priemer: {2:.4f}".format(min, max, priemer))

# vypočítame min, max a priemernú hodnotu pre druhý stĺpec (weight)
suma = 0
min = None
max = None
for riadok in riadky:
    hodnota = riadok[1]
    if hodnota is None: # preskočíme chýbajúce hodnoty  
        continue
    if min is None or hodnota < min:  # ak sme ešte nepriradili do min žiadnu hodnotu, alebo ak je hodnota v riadku menšia než minimum, priradíme ju do min
        min = hodnota
    if max is None or hodnota > max:  # podobne ak sme ešte nepriradili max, alebo je hodnota väčšia než max, priradíme ju do max 
        max = hodnota
    suma += hodnota # pripočítame hodnotu do sumy pre priemer
priemer = suma / len(riadky)

print("weight min: {0:.4f}, max: {1:.4f}, priemer: {2:.4f}".format(min, max, priemer))

# pre textový stĺpec producer zozbierame zoznam hodnôt
hodnoty = []
for riadok in riadky:
    hodnoty.append(riadok[3]) # štvrtý stĺpec
hodnoty.sort() # usporiadame hodnoty
print(hodnoty)
```
